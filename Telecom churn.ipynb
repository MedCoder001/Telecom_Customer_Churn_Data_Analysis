## 3. Clean data

```python
# Remove underscores from column names
dataset1.columns = dataset1.columns.str.replace('_', '')
```

```python
# Renaming the column names in dataset1 to match that of dataset2 by defining a dictionary to map original column names to new names
new_column_names = {
    'married': 'partner',
    'numberofdependents': 'dependents',
    'tenureinmonths': 'tenure',
    'deviceprotectionplan': 'deviceprotection',
    'premiumtechsupport': 'techsupport',
    'monthlycharge': 'monthlycharges',
    'customerstatus': 'churn'
}

# Use the rename() method to rename the columns
dataset1.rename(columns=new_column_names, inplace=True)
```

```python
# Rearrange the columns in dataset1 to match the order of columns in dataset2
# Define the desired column order
desired_column_order = ['customerid', 'gender', 'partner', 'dependents', 
                        'tenure', 'phoneservice', 'multiplelines', 'internetservice', 'onlinesecurity', 
                        'onlinebackup', 'deviceprotection', 'techsupport', 'streamingtv', 'streamingmovies', 
                        'contract', 'paperlessbilling', 'paymentmethod', 'monthlycharges', 'totalcharges', 'churn']

# Create a new DataFrame with the desired column order
dataset1 = dataset1[desired_column_order]
```

### Quality Issue 1: Validity

The "internetservice" column in the dataset contains values like "Fiber optic" and "DSL," which can be replaced with a more generalized "Yes" value to simplify analysis and interpretation.

```python
# Count the occurrences in the "internetservice" column
counts = dataset2['internetservice'].value_counts()
counts
```

To address this issue, we can replace "Fiber optic" and "DSL" with "Yes" in the "internetservice" column. After replacement, we can count the occurrences again to ensure that the values have been updated:

```python
# Replace "Fiber optic" and "DSL" with "Yes" in the "internetservice" column
dataset2['internetservice'].replace(['Fiber optic', 'DSL'], 'Yes', inplace=True)

# Count the occurrences after replacement
counts = dataset2['internetservice'].value_counts()

# Display the updated counts
print(counts)
```

```python
# Define a function to map values to 'No' or 'Yes'
def map_dependents(value):
    if value == '0':
        return 'No'
    else:
        return 'Yes'

# Apply the mapping function to the "dependents" column
dataset1['dependents'] = dataset1['dependents'].apply(map_dependents)
```

### Quality Issue 2: Churn column values

In the "Churn" column of dataset1, the values "Stayed" are replaced with "No" and "Churned" are replaced with "Yes." This is a data cleaning step to ensure that the "Churn" column contains consistent and clear values.

```python
# Replace "Stayed" with "No" and "Churned" with "Yes" in the "Churn" column
dataset1['churn'].replace('Stayed', 'No', inplace=True)
dataset1['churn'].replace('Churned', 'Yes', inplace=True)

# Count the occurrences after replacement
counts = dataset1['churn'].value_counts()

# Display the updated counts
print(counts)
```

## 5. Answering the research questions

```python
# Set the style for Seaborn plots
sns.set(style="whitegrid")
```

```python
# Load the dataset from the CSV file
merged_datasets = pd.read_csv('merged_dataset.csv') 
```

### **5.1.** Research question 1: How are the categorical variables contributing to customer churn?

```python
# Visualizations for categorical columns
categorical_columns = ['gender', 'partner', 'dependents', 'phoneservice', 'multiplelines', 
                       'internetservice', 'onlinesecurity', 'onlinebackup', 'deviceprotection', 
                       'techsupport', 'streamingtv', 'streamingmovies', 'contract', 
                       'paperlessbilling', 'paymentmethod', 'churn']

# Visualizations for binary categorical columns (e.g., 'gender', 'partner', 'dependents')
plt.figure(figsize=(15, 15))
for i, col in enumerate(categorical_columns, 1):
    plt.subplot(4, 4, i)
    sns.countplot(data=merged_datasets, x=col)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Count', fontsize=12)

plt.tight_layout()
plt.show()
```

Categorical variables such as Internet Service, Online Security, and Online Backup were found to be significant predictors of customer churn. Customers with Fiber Optic Internet Service, No Online Security, and No Online Backup were more likely to churn. StreamingTV and StreamingMovies have a minor effect on customer churn.

### **5.2.** Research question 2: How are the numerical variables contributing to customer churn?

```python
# Visualizations for numeric columns (e.g., 'tenure', 'monthlycharges', 'totalcharges')
numeric_columns = ['tenure', 'monthlycharges', 'totalcharges']

plt.figure(figsize=(15, 5))
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(1, 3, i)
    sns.histplot(data=merged_datasets, x=col, kde=True)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)

plt.tight_layout()
plt.show()
```

The findings indicated that customers with higher Monthly Charges were more likely to churn, while those with longer Tenure and higher Total Charges were less likely to churn. 

### **5.3.** Research question 3: How does contract type affect churn rate?

```python
# Churn rate by contract type
churn_rate_by_contract = merged_datasets.groupby('contract')['churn'].value_counts(normalize=True).unstack(fill_value=0)['Yes'].reset_index()

# Create the bar plot
plt.figure(figsize=(8, 6))
sns.barplot(data=churn_rate_by_contract, x='contract', y='Yes')
plt.xlabel('Contract Type', fontsize=12)
plt.ylabel('Churn Rate', fontsize=12)
plt.title('Churn Rate by Contract Type', fontsize=14)
plt.show()
```

The analysis demonstrated that customers with Month-to-Month contracts had a significantly higher churn rate compared to those with longer-term contracts. 

### **5.4.** Research question 4: How does payment method affect churn rate?

```python
# Churn rate by payment method
churn_rate_by_payment = merged_datasets.groupby('paymentmethod')['churn'].value_counts(normalize=True).unstack(fill_value=0)['Yes'].reset_index()

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=churn_rate_by_payment, x='paymentmethod', y='Yes')
plt.xlabel('Payment Method', fontsize=12)
plt.ylabel('Churn Rate', fontsize=12)
plt.title('Churn Rate by Payment Method', fontsize=14)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

The results showed that customers using Electronic Check and Mailed Check as their payment method had a notably higher churn rate, while those using Bank Transfer (Auto) and Credit Card (Auto) had a lower churn rate.
